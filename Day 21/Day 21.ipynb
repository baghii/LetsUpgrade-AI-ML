```python
import pandas as pd
import statsmodels.api as sm
```

```python
dataset = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name="Data")
```

```python
dataset.head()
```

```python
x = dataset.Personal_Loan
```

```python
y = dataset[['Age', 'Experience','Income','CCAvg','Education','Mortgage','Securities_Account','CD_Account','Online','CreditCard']]
```

```python
y.head()
```

```python
y1 = sm.add_constant(y)
```

```python
logistic = sm.Logit(x,y1)
```

```python
result = logistic.fit()
```

```python
result.summary()
```
